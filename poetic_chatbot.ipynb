```python
import time
from google import genai
from google.colab import userdata
```

**3- Retrieving the API key**

```python
key = userdata.get('GEMINI_API_KEY')
```

**Building the chatbot**

**1- add a text streaming function**

```python
def stream (text):
  print('Bot:', end='')
  words = text.split()
  for word in words:
    print(word, end='', flush=True)
    time.sleep(1)
  print()
```

**2- connect to gemini API**

```python
try:
  client = genai.Client(api_key=key)
except KeyError:
  print(f'Error with gemini api key.\n error message:\n{KeyError}')
chat = client.chats.create(model="gemini-2.0-flash")
```

**3- create a function to talk to gemini**

```python
def ask_gemini(user_input):
    try:
        prompt = f"You are a poetic companion.For every user message, respond with short, original poetic verses. Use vivid imagery, emotion, and gentle rhyme when it fits. Make each reply feel personal to what the user said. Write 2–4 lines per reply, separated by line breaks. Do not explain, just answer in poetry."
        response = chat.send_message(user_input)
        return response.text
    except Exception as e:
        if "429" in str(e):
            return "I'm temporarily out of quota. Please wait and try again later."
        return f"Oops, something went wrong: {e}"
```

**4- build the chatbot loop**

```python
print("Welcome to your Gemini Chatbot! Type 'quit' to exit.")
while True:
  user_input = input("You: ")
  if user_input.lower() == 'quit':
    print("Bot: Bye for now!")
    break
  response = ask_gemini(user_input)
  stream(response)
```
